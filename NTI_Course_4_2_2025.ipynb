```python
canada['date'] = pd.to_datetime(canada['date'])
```

**Total Number of Reported Cases and Deaths**

```python
total_cases = canada['total_cases'].max()
total_deaths = canada['total_deaths'].max()
print(f"Total Cases: {total_cases}, Total Deaths: {total_deaths}")
```

**Total Number of Tests Conducted**

```python
total_tests = canada['total_tests'].max()
print(f"Total Tests : {total_tests}")
```

**When was the deadliest day in Canada?**

```python
peak_deaths = canada.loc[canada['new_deaths'].idxmax()]
print(peak_deaths[['date', 'new_deaths']])
```

**When was the worst wave of COVID-19 in Canada?**

```python
peak_cases = canada.loc[canada['new_cases'].idxmax()]
print(peak_cases[['date', 'new_cases']])
```

**When were hospitals under the most pressure?**

```python
peak_hosp = canada.loc[canada['hosp_patients'].idxmax()]
print(peak_hosp[['date', 'hosp_patients']])
```

**What percentage of people recovered from COVID-19?**

```python
recovery_rate = ((canada['total_cases'] - canada['total_deaths']) / canada['total_cases']) * 100
print(f"Recovery Rate: {recovery_rate.max():.2f}%")
```

**Daily New Cases Over Time**

```python
plt.figure(figsize=(12,6))
plt.plot(canada['date'], canada['new_cases'], label='New Cases', color='blue')
plt.xlabel("Date")
plt.ylabel("New Cases")
plt.title("Daily COVID-19 Cases in Canada")
plt.legend()
plt.show()
```

**Daily New Deaths Over Time**

```python
plt.figure(figsize=(12,6))
plt.plot(canada['date'], canada['new_deaths'], label='New Deaths', color='red')
plt.xlabel("Date")
plt.ylabel("New Deaths")
plt.title("Daily COVID-19 Deaths in Canada")
plt.legend()
plt.show()
```

**7-day Moving Average for Cases**

```python
canada['new_cases_7d_avg'] = canada['new_cases'].rolling(window=7).mean()

plt.figure(figsize=(12,6))
plt.plot(canada['date'], canada['new_cases_7d_avg'], label='7-day Avg New Cases', color='green')
plt.xlabel("Date")
plt.ylabel("New Cases (7-day Avg)")
plt.title("7-day Moving Average of COVID-19 Cases in Canada")
plt.legend()
plt.show()
```

**Peak Days for New Cases and Deaths**

**Fully Vaccinated Population**

**Testing Trends**

```python
plt.figure(figsize=(12,6))
plt.plot(canada['date'], canada['total_tests'], label='Total Tests', color='orange')
plt.xlabel("Date")
plt.ylabel("Total Tests")
plt.title("Total COVID-19 Tests Conducted in Canada")
plt.legend()
plt.show()
```

**Correlation Heatmap**

```python
correlation_matrix = canada[['new_cases', 'new_deaths', 'people_vaccinated', 'total_cases', 'total_deaths']].corr()

plt.figure(figsize=(8,6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Heatmap")
plt.show()
```

**Cases vs. Deaths Scatter Plot**

```python
plt.figure(figsize=(8,6))
plt.scatter(canada['total_cases'], canada['total_deaths'])
plt.xlabel("Total Cases")
plt.ylabel("Total Deaths")
plt.title("COVID-19 Cases vs. Deaths in Canada")
plt.show()
```

**Did the case growth rate slow down over time?**

```python
canada['case_growth_rate'] = canada['new_cases'].pct_change() * 100

plt.figure(figsize=(12,6))
plt.plot(canada['date'], canada['case_growth_rate'], color='purple', label='Case Growth Rate (%)')
plt.xlabel("Date")
plt.ylabel("Growth Rate (%)")
plt.title("COVID-19 Case Growth Rate in Canada")
plt.legend()
plt.show()
```

**Was there a second wave?**

```python
plt.figure(figsize=(12,6))
plt.plot(canada['date'], canada['new_cases'], label="New Cases")
plt.axvline(canada['date'].iloc[canada['new_cases'].idxmax()], color='r', linestyle='--', label="Peak")
plt.xlabel("Date")
plt.ylabel("Cases")
plt.title("Second Wave Analysis")
plt.legend()
plt.show()
```

**How did hospitalizations compare to cases?**

```python
# Define major waves manually based on data insights
wave_1 = canada[(canada['date'] >= '2020-03-01') & (canada['date'] <= '2020-07-01')]['total_deaths'].max()
wave_2 = canada[(canada['date'] >= '2020-10-01') & (canada['date'] <= '2021-02-01')]['total_deaths'].max()
wave_3 = canada[(canada['date'] >= '2021-06-01') & (canada['date'] <= '2021-10-01')]['total_deaths'].max()
wave_4 = canada[(canada['date'] >= '2021-12-01') & (canada['date'] <= '2022-04-01')]['total_deaths'].max()

values = [wave_1, wave_2, wave_3, wave_4]
labels = ['First Wave', 'Second Wave', 'Third Wave', 'Fourth Wave']

fig = px.pie(names=labels, values=values,
             title="COVID-19 Deaths Across Major Waves in Canada",
             template="plotly_dark")

fig.update_traces(textinfo='percent+label', pull=[0.1, 0.1, 0, 0])
fig.show()
```
